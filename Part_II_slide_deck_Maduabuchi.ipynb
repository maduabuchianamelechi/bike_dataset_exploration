## Dataset Overview

> The Bike Dataset contains information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
df = pd.read_csv('San_francisco_bay_area_bike_data_201902.csv')
bike_df = df.copy()
````

## Distribution of the Trip Duration

> After converting the duration column from seconds (secs) to minutes (mins) and applying a log transformation. The `duration` column in minutes have a unimodal distribution with a peak at the 10th minute showing that a rider spends an average of 10mins on their ride.

````python
bike_df["duration_min"] = bike_df["duration_sec"] / 60
bike_df.drop("duration_sec" , axis = 1 , inplace = True)

binedges = 10**np.arange(0 , 3+0.1 , 0.1)
ticks =  [0.1, 0.3, 1, 3, 10, 30, 100, 300]
ax = plt.axes()
plt.hist(data = bike_df , x = 'duration_min' , bins = binedges);
plt.xscale('log')
ax.set_xticks(ticks)
ax.set_xticklabels(['0.1', '0.3', '1', '3', '10', '30', '100', '300'])
ax.set_title('Distribution of Ride Duration', fontsize=15, weight='bold')
ax.set_xlabel(' Ride Duration (Minutes)', fontsize=10, weight='bold')
ax.set_ylabel('Count', fontsize=10, weight='bold')

plt.show();
````

## Distribution of Customer Type

> We also examined the type of customers we have, it has been revealed that we have more subscribers than casual customers. The subcribers makes up about `89%` of our customer base

````python
bin_color = sb.color_palette()[0]
plt.figure(figsize=[8,6])
plot_df = bike_df['user_type'].value_counts(normalize=True)*100
plot_df.plot(kind='bar', rot=0, color=bin_color)
plt.xlabel('Customer Type', fontsize=10, weight='bold')
plt.ylabel('Percentage', fontsize=10, weight='bold')
plt.title('Distribution of Customer Type', fontsize=15, weight='bold');
````

## Customer Type vs Trip Duration

> Even though our customer base is made up of about 89% subscribers, further analysis revealed that casual customers have the tendency of renting the bike for a longer period so the business should strategize on how to grow our casual customer base.

````python
plt.figure(figsize=[8,6])
sb.boxplot(data = bike_df, x = 'user_type', y = 'duration_min')
plt.ylim(0, 100)
plt.xlabel('Customer Type', fontsize=10, weight='bold')
plt.ylabel('Duration (minutes)', fontsize=10, weight='bold')
plt.title('Relationship between Trip Duration and Customer Type', fontsize=14, weight='bold');
````

## Customer Type vs Gender

> Taking a look at the distribution of our customers based on gender, we can see that the male gender makes up a higher percentage of our customer base.

````python
plt.figure(figsize=[8,6])
sb.countplot(data = bike_df, x = 'member_gender', hue = 'user_type')
plt.xlabel('Gender', fontsize=10, weight='bold')
plt.ylabel('Count', fontsize=10, weight='bold')
plt.title('Distribution of Gender vs Customer Type', fontsize=15, weight='bold')
plt.legend(title='Customer Type');
````
